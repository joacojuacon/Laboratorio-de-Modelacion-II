```python
porcentaje_faltantes = (df_numerico.isnull().sum() / len(df_numerico)) * 100

print(porcentaje_faltantes)
```

```python
for columna in df_numerico.columns:
    
    # Calculamos el porcentaje de faltantes solo para esta columna
    porcentaje = (df_numerico[columna].isnull().sum() / len(df_numerico)) * 100
    
    # Aquí está tu 'if'. 
    # (Si decides seguir la regla estadística clásica, solo cambia el '>' por un '<' o '<=')
    if porcentaje < 15:
        # Si cumple, imputamos con la mediana
        mediana_columna = df_numerico[columna].median()
        df_numerico[columna] = df_numerico[columna].fillna(mediana_columna)
        
        # Un print opcional para saber qué hizo el código
        print(f"Columna '{columna}' imputada (Tenía {porcentaje:.1f}% de faltantes).")
        
    else:
        # El famoso "nain". No hace nada, pero lo ponemos visualmente 
        # para que veas que la columna se ignoró.
        pass 

print("\n¡Proceso terminado!")
```

```python
# 1. Filtramos donde el cobre NO es nulo
df_con_cobre = df[df['Cu_ppm'].notna()].copy() # Agregamos .copy() para evitar advertencias

# 2. Convertimos la columna a string por si acaso hay problemas de tipo de dato
df_con_cobre['LITOLOGIA'] = df_con_cobre['LITOLOGIA'].astype(str)

# 3. Calculamos los conteos
conteos = df_con_cobre['LITOLOGIA'].value_counts()

# 4. CRUCIAL: Convertimos el índice explícitamente a una lista de strings
categorias_grandes = conteos[conteos > 5].index.astype(str).tolist()

# 5. Volvemos a intentar el filtro
df_cobre = df_con_cobre[df_con_cobre['LITOLOGIA'].isin(categorias_grandes)]

# Verificación
print(f"Filas en df_cobre: {len(df_cobre)}")
```

```python
df_con_cobre['LITOLOGIA'].isin(categorias_grandes)
```

```python
df_cobre = df_con_cobre[df_con_cobre['LITOLOGIA'].isin(categorias_grandes)]
```

```python
len(df_cobre)
```

```python
grupos = [df_cobre['Cu_ppm'][df_cobre['LITOLOGIA'] == lito] for lito in df_cobre['LITOLOGIA'].unique()]
# Opción B: Pedirle a la prueba de scipy que ignore los NaN automáticamente
estadistico, p_valor = kruskal(*grupos, nan_policy='omit')
```

```python
print(p_valor)
```

```python
print(estadistico)
```

```python
H = estadistico
k = df_cobre['LITOLOGIA'].nunique()
n = len(df_cobre)
```

```python
eta_cuadrado = (H - k + 1) / (n - k)
```

```python
print(f"Tamaño del efecto (Eta-cuadrado): {eta_cuadrado:.4f}")
```

```python
import scikit_posthocs as sp
```

```python
cobre = sp.posthoc_dunn(df_cobre, val_col='Cu_ppm', group_col='LITOLOGIA', p_adjust='holm')
```
